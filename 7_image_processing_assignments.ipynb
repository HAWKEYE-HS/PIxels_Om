## Rotating an image by 180 degrees

````python
import matplotlib.pyplot as plt
import numpy as np
import math
from PIL import Image


def rotate(image, angle):
    rads = math.radians(angle)
    # Center point of the original image
    cx, cy = (image.shape[1] // 2, image.shape[0] // 2)

    # Height and Width of the rotated image
    height_rot_img = round(abs(image.shape[0] * math.sin(rads))) + round(abs(image.shape[1] * math.cos(rads)))
    width_rot_img = round(abs(image.shape[1] * math.cos(rads))) + round(abs(image.shape[0] * math.sin(rads)))
    rot_img = np.uint8(np.zeros((height_rot_img, width_rot_img, image.shape[2])))
    # Center point of rotated image
    midx, midy = (width_rot_img // 2, height_rot_img // 2)

    for i in range(rot_img.shape[0]):
        for j in range(rot_img.shape[1]):
            x = (i - midx) * math.cos(rads) + (j - midy) * math.sin(rads)
            y = -(i - midx) * math.sin(rads) + (j - midy) * math.cos(rads)
            x = round(x) + cy
            y = round(y) + cx

            if x >= 0 and y >= 0 and x < image.shape[0] and y < image.shape[1]:
                rot_img[i, j, :] = image[x, y, :]
    return rot_img


if __name__ == '__main__':
    image = np.array(Image.open((r"images\blur.jpeg")))
    rotated_image = rotate(image, 180)
    plt.imshow(rotated_image)
    plt.show()
````

# **Convolution Tasks**

***Reference*** code : 

````python
#############Suboptimal implementation of convolution #####################
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convolve(kernel, orig_img):
    #####################Flip kernel suboptimally####################
    kernel_inv = np.copy(kernel)
    kernel_inv[0] = kernel[2]
    kernel_inv[2] = kernel[0]
    kernel_inv[:, 0] = kernel_inv[:, 2]
    kernel_inv[:, 2] = kernel_inv[:, 0]
    #################################################################
    img = cv2.copyMakeBorder(orig_img, 1, 1, 1, 1, cv2.BORDER_CONSTANT) # Add padding
    res_img = orig_img.copy()
    for i in range(1, img.shape[0] - 1):
        for j in range(1, img.shape[1] - 1):
                res_img[i - 1, j - 1] = np.sum(img[i - 1 : i + 2, j - 1 : j + 2] * kernel_inv[:, :])
    return res_img
  
orig_img = cv2.imread("./Assets/dog_test.png", 0) # Read Image
kernel = np.array([[ 1,  1,  1], [ 0,  0,  0], [-1, -1, -1]]) # Simple vertical gradient kernel

res = convolve(kernel, orig_img)

orig = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)
plt.imshow(orig)
plt.show()
plt.imshow(res, cmap = plt.cm.gray)
plt.show()
````

## Q2. Apply Gaussian smoothing using normal 2D convolution.

````python
import cv2
import numpy as np


def gauss_function_2D(k):
    radius = (k - 1) / 2
    sigma = radius / 2.

    kernel = np.arange(2 * radius + 1)
    row = np.exp(-(((kernel - radius) / (sigma)) ** 2) / 2.)
    col = row.transpose()
    out = np.outer(row, col)
    out = out / np.sum(out)
    return out
def GaussianBlur2D_grayscale(img_path, k):
    kernel = gauss_function_2D(k)
    print(kernel)
    image = cv2.imread(img_path, 0)
    image2 = np.pad(image, (1, 1), mode='constant')
    blurred_image = np.copy(image2)
    m, n = kernel.shape[:]
    y, x = image2.shape[:2]

    y1 = y - m + 1
    x1 = x - n + 1

    for i in range(y1):
        for j in range(x1):
            blurred_image[i][j] = np.sum(image2[i:i + m, j:j + n] * kernel)

    cv2.imshow("Blurred_Image", blurred_image)
    cv2.imshow("Original", image)
    cv2.waitKey(0)
def GaussianBlur2D_color(img_path, k):
    kernel = gauss_function_2D(k)
    print(kernel)
    image = cv2.imread(img_path)
    blank = np.zeros(image.shape[:2], dtype='uint8')
    image = image[:, :, 0]
    image2 = np.pad(image, (1, 1), mode='constant')
    blue_image = np.copy(image2)
    m, n = kernel.shape[:]
    y, x = image2.shape[:2]

    y1 = y - m + 1
    x1 = x - n + 1

    for i in range(y1):
        for j in range(x1):
            blue_image[i][j] = np.sum(image2[i:i + m, j:j + n] * kernel)

    image = cv2.imread(img_path)
    image = image[:, :, 1]
    image2 = np.pad(image, (1, 1), mode='constant')
    green_image = np.copy(image2)
    m, n = kernel.shape[:]
    y, x = image2.shape[:2]

    y1 = y - m + 1
    x1 = x - n + 1

    for i in range(y1):
        for j in range(x1):
            green_image[i][j] = np.sum(image2[i:i + m, j:j + n] * kernel)

    image = cv2.imread(img_path)
    image = image[:, :, 2]
    image2 = np.pad(image, (1, 1), mode='constant')
    red_image = np.copy(image2)
    m, n = kernel.shape[:]
    y, x = image2.shape[:2]

    y1 = y - m + 1
    x1 = x - n + 1

    for i in range(y1):
        for j in range(x1):
            red_image[i][j] = np.sum(image2[i:i + m, j:j + n] * kernel)

    blurred_image=cv2.merge([blue_image,green_image,red_image])
    cv2.imshow("Blurred_Image", blurred_image)
    cv2.waitKey(0)
if __name__ == '__main__':
    GaussianBlur2D_color(r"C:\Users\omdoi\OneDrive\Pictures\BackgroundImages\RDJ.jpg",5)
````

````python
def GaussianBlur2D_separable(img_path, k):
    kernel = gauss_function_2D(k)
    E_index = kernel.nonzero()[0][0], kernel.nonzero()[1][0]
    kernel_col = kernel[:, E_index[1]].reshape(-1, 1)
    kernel_row = kernel[E_index[0], :].reshape(1, -1)
    print(kernel_col)
    print(kernel_row)
    image = cv2.imread(img_path, 0)
    image2 = np.pad(image, (1, 1), mode='constant')
    blurred_image = np.copy(image2)
    m1, n1 = kernel_col.shape[:]
    y, x = image2.shape[:2]
    m2, n2 = kernel_row.shape[:]
    y1 = y - m1 + 1
    x1 = x - n1 + 1
    y2 = y - m2 + 1
    x2 = x - n2 + 1
    for i in range(y2):
        for j in range(x2):
            blurred_image[i][j] = np.sum(image2[i:i+m2, j:j + n2] * kernel_col)
    for i in range(y1):
        for j in range(x1):
            blurred_image[i][j] += np.sum(image2[i:i + m1, j:j+n1] * kernel_row)
    cv2.imshow("Blurred_Image", blurred_image)
    cv2.imshow("Original", image)
    cv2.waitKey(0)
````

## TASK 2
Playing With Contours
* Refer [OpenCV docs](https://docs.opencv.org/3.4/d4/d73/tutorial_py_contours_begin.html) and study the types of contours
* Task is to find 4 points i.e. leftmost rightmost top and bottom point as shown in the image below.

**INPUT**
![](./images/contour_input.jpeg)

**OUTPUT**
![](./images/contour_output1.jpeg)
![](./images/contour_output2.jpeg)

````python
import cv2
import imutils   # A module which consists of a series of convenience functions for making image processing tasks such as rotation, resizing, detecting edges, etc easier with OpenCV
image = cv2.imread(r"C:\Users\omdoi\OneDrive\Pictures\BackgroundImages\contour_input.jpeg")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges = cv2.threshold(gray_image, 54, 255, cv2.THRESH_BINARY)[1]
# Eroding and dilating the image to remove unwanted noise
edges = cv2.erode(edges, None)
edges = cv2.dilate(edges, None)
contours = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(contours)
# Finding the largest contour
largest_contour = max(contours, key=cv2.contourArea)
# Determining the extreme points
Left = tuple(largest_contour[largest_contour[:, :, 0].argmin()][0])
Top = tuple(largest_contour[largest_contour[:, :, 1].argmin()][0])
Right = tuple(largest_contour[largest_contour[:, :, 0].argmax()][0])
Bottom = tuple(largest_contour[largest_contour[:, :, 1].argmax()][0])
cv2.drawContours(image, [largest_contour], -1, (0, 255, 255), 2)
# Drawing circles around those extreme points
cv2.circle(image, Left, 8, (0, 0, 255), -1)
cv2.circle(image, Right, 8, (0, 255, 0), -1)
cv2.circle(image, Top, 8, (255, 0, 0), -1)
cv2.circle(image, Bottom, 8, (255, 255, 0), -1)
cv2.imshow("Image", image)
cv2.waitKey(0)

````
